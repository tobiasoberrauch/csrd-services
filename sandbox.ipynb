```python
from csrd_services.core.esrs_classification import ESRS_CATEGORIES


transformed = []

for full_key, themes in ESRS_CATEGORIES.items():
    if ":" in full_key:
        category, description = full_key.split(":", 1)
        
        processed_themes = []
        for theme in themes:
            if ":" in theme:
                theme_category, theme_description = theme.split(":", 1)
                processed_themes.append({
                    "category": theme_category.strip(),
                    "description": theme_description.strip()
                })
            else:
                processed_themes.append({
                    "category": theme,
                    "description": ""
                })
                
        transformed.append({
            "category": category.strip(),
            "description": description.strip(),
            "themes": processed_themes
        })

transformed
```

```python

# Ausgabe
import json

# save in json file
def save_json(path, data):
    with open(path, 'w', encoding='utf-8') as f:
        json.dump(data, f, indent=4, ensure_ascii=False)


save_json("esrs_categories.json", transformed)

#
```
